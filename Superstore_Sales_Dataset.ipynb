# **Venda por ano**

```python
# Criar uma coluna com ano
base["Order Date"] = pd.to_datetime(base["Order Date"], format="%d/%m/%Y")
base['Ano'] = base['Order Date'].dt.year
```

```python
base.groupby("Ano")["Sales"].sum().plot.bar();
```

```python
import matplotlib.pyplot as plt
import numpy as pd
```

```python
# make data
x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * x)

# plot
fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2.0)

ax.set(xlim=(0,8), xticks=np.arange(0, 8),
       ylim=(0,8), yticks=np.arange(0, 8))

plt.show()
```

```python
# plot
fig, ax = plt.subplots(figsize=(8, 5))

soma_ano = base.groupby("Ano")["Sales"].sum()

ax.bar(soma_ano.index, soma_ano.values, linewidth=2.0, color='#84ba5b')

ax.set_title("VENDAS POR ANO",
             fontsize= 12,
             fontweight="bold",
             color= "#84ba5b"
             )

ax.xaxis.set_ticks([2015, 2016, 2017, 2018])
ax.tick_params(axis='x', labelsize=12)

for i in range(0,4):
  ax.annotate('{:,.0f}'.format(soma_ano.values[i]).replace(",", "."),
              (soma_ano.index[i],soma_ano.values[i]),
              ha="center",
              xytext=(0,5),
              textcoords="offset points",
              fontsize=12,
              color="#84ba5b",
              fontweight="bold"
              )

ax.set_yticks([0, 790000])
ax.yaxis.set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

plt.show()
```

# **Venda por mês**

```python
# Criando uma coluna com o mês
base['Mes'] = base['Order Date'].dt.month
```

```python
base.head(2)
```

```python
# Fazendo o groupby agora por mês e ano
soma_mes = base.groupby(["Mes", "Ano"])["Sales"].sum()
```

```python
soma_mes.head()
```

```python
# Transformar os índices em colunas
soma_mes = soma_mes.reset_index()
```

```python
soma_mes.head()
```

```python
soma_mes[soma_mes["Ano"] == 2015].Sales.values
```

```python
# Criando o gráfico de mês com o matplotlib
fig,ax = plt.subplots(
    figsize = (9,4)
)

wid = 0.2

ax.bar(soma_mes[soma_mes.Ano == 2015].Mes.values-wid,
       soma_mes[soma_mes.Ano == 2015].Sales.values,
       color="#84ba5b",
       label=2015,
       width=wid
       )
ax.bar(soma_mes[soma_mes.Ano == 2016].Mes.values+wid/2,
       soma_mes[soma_mes.Ano == 2016].Sales.values,
       color="#00a9b5",
       label=2016,
       width=wid
       )
ax.bar(soma_mes[soma_mes.Ano == 2017].Mes.values+wid/2,
       soma_mes[soma_mes.Ano == 2017].Sales.values,
       color="#1b618c",
       label=2016,
       width=wid
       )
ax.bar(soma_mes[soma_mes.Ano == 2018].Mes.values+wid/2,
       soma_mes[soma_mes.Ano == 2018].Sales.values,
       color="#177013",
       label=2016,
       width=wid
       )

ax.legend()

plt.show()
```
