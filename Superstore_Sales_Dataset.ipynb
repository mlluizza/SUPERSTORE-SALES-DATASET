### **Juntando bases no pandas**

```python
# Criando 2 dataframes
import pandas as pd
dic1 = {
    "nomes" : ['Nome1', 'Nome2', 'Nome3'],
    "valores": [1,2,3]
}

base_dic1 = pd.DataFrame(dic1)

dic2 = {
    "nomes" : ['Nome1', 'Nome2', 'Nome3'],
    "valores" : [9,8,7]
}

base_dic2 = pd.DataFrame(dic2)
```

```python
base_merge = pd.merge(
    base_dic1, # Primeira base
    base_dic2,  # Segunda base
    on="nomes",  # Tipo de junção
    how="outer"   # Coluna junção de bases
)

display(base_merge)
```

#### **Como podemos cruzar a base com todos os anos de 2015 a 2018?**

```python
df_datas = pd.DataFrame([2018, 2017, 2016, 2015])
df_datas.columns = ['Ano']
df_datas
```



*   Então podemos, nas duas bases, criar uma coluna chamada união com o mesmo valor



```python
df_datas['uniao'] = 'unir'
itens['uniao'] = 'unir'
```

```python
# Definindo o top n itens
n = 10
top_n_itens = itens.head(n)
```

```python
base_merge = pd.merge(
    top_n_itens, # Primeira base, alterada para top_n_itens
    df_datas,  # Segunda base, alterada para df_datas
    on="uniao",  # Coluna de junção
    how="outer"   # Tipo de junção
)

display(base_merge) # Alterado para exibir o resultado da junção
```

```python
itens_ano_grafico = pd.merge(
    base_merge,
    itens_ano,
    how = 'left',
    on=['Product Name', 'Ano'] # Corrected syntax
)

display (itens_ano_grafico)
```
